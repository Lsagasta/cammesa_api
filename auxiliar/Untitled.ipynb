```python
import pandas as pd

df = pd.read_csv("archivo_unificado.csv")

df
```

```python
df.rename(columns={"fecha":"timestamp"},inplace = True)
```

```python
import datetime
```

```python
df['fecha'] = pd.to_datetime(df['timestamp']).dt.date
df["hora"] = pd.to_datetime(df["timestamp"]).dt.strftime('%H:%M:%S')

dff = df[["timestamp","sumTotal","hidraulico","termico","nuclear","renovable","importacion","fecha","hora"]]

dff

dff.to_csv("unificado_final",index = False)
```
